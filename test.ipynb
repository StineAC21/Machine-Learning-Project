```python
plt.figure(figsize=(10, 6))
df['score'].hist(bins=[i * 0.5 for i in range(21)])
plt.title('Distribution of Movie Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot histograms of scores by genre
plt.figure(figsize=(14, 10))
for i, genre in enumerate(filtered_genres):
    plt.subplot(3, 2, i + 1)
    sns.histplot(df[df['genre'] == genre]['score'].dropna(), bins=[i * 0.5 for i in range(21)], kde=True)
    plt.title(f'Score Distribution for {genre}')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Group by year and calculate the mean score
average_score_by_year = df.groupby('year')['score'].mean()

# Plot the average score by year
plt.figure(figsize=(10, 6))
average_score_by_year.plot(kind='line')
plt.title('Average Movie Score by Year')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.grid(True)
plt.show()
```
